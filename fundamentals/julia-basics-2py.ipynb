# Plotting

```python
import matplotlib.pyplot as plt
```

```python
# Create an array of 1000 points from 0 to 2*pi
x = np.linspace(0, 2*np.pi, 1000)

# Calculate y as the sine of 3x
y = np.sin(3 * x)

# Plotting
plt.plot(x, y, color="red", linewidth=2.0, linestyle="--")
plt.title("The sine function")
plt.show()
```

```python
# Create an array of 1000 points from 0 to 2*pi
x = np.linspace(0, 2 * np.pi, 1000)

# Calculate y as the sine of 3x
y = np.sin(3 * x)

# Calculate z as the cosine of x
z = np.cos(x)

# Plotting sin(3x)
plt.plot(x, y, color="red", linewidth=2.0, linestyle="--", label="sin(3x)")

# Plotting cos(x)
plt.plot(x, z, color="blue", linewidth=1.0, linestyle="-", label="cos(x)")

# Adding legend
plt.legend(loc="upper center")

# Show the plot
plt.show()
```
